```python
import pandas as pd
from bs4 import BeautifulSoup
import csv
import requests as req
from collections import Counter
```

# Keyword Categorisation: Target type 

```python
#list of target keywords with replicates with different cases 
target_types=['protease','nuclease','phospholipase','deaminase','methyltransferase','phosphotransferase','protein-protein','ATPase','synthetase','hydrolase','endoribonuclease','helicase','kinase','aminopeptidase','ligase','lyase','receptor','GTPase','regulator','epigenetic','cation','oxidoreductase','transporter','transcription',
'proteases','nucleases','phospholipases','deaminases','methyltransferases','phosphotransferases','ATPases','synthetases','hydrolases','endoribonucleases','helicases','kinases','aminopeptidases','ligases','lyases','receptors','GTPases','regulators','epigenetics','cations','oxidoreductases','transporters','transcriptions','Protease','Nuclease','Phospholipase','Deaminase','Methyltransferase','Phosphotransferase','Protein-Protein','Synthetase','Hydrolase','Endoribonuclease','Helicase','Kinase','Aminopeptidase','Ligase','Lyase','Receptor','Regulator','Epigenetic','Cation','Oxidoreductase','Transporter','Transcription',
'Proteases','Nucleases','Phospholipases','Deaminases','Methyltransferases','Phosphotransferases','Synthetases','Hydrolases','Endoribonucleases','Helicases','Kinases','Aminopeptidases','Ligases','Lyases','Receptors','Regulators','Epigenetics','Cations','Oxidoreductases','Transporters','Transcriptions']
#save to csv file 
with open("Target_type_keywords.csv", mode='w', newline='') as csv_file:
    # Create a CSV writer
    csv_writer = csv.writer(csv_file)
    
    # Loop through the list and write each string as a row
    for item in target_types:
        csv_writer.writerow([item])
```

```python
def target_type(aid):
    """
    categorises assays by keyword matching from xml file
    
    aid: str PubChem Bioassay unique identifier 
    
    return: category label
    """
    string='https://pubchem.ncbi.nlm.nih.gov/rest/pug_view/data/assay/'+aid+'/XML' #URL string 
    # Requesting for the website
    Web = req.get(string,verify=False) 
    # Creating a BeautifulSoup object and specifying the parser
    S = BeautifulSoup(Web.text,features="xml")
    string_counter = Counter() # Counter for counting keyword occurrences 
    for string in target_types:#iterate through each keyword of each target category 
        if string in S.get_text(): #check if keyword is in lowercase text of xml file 
            string_counter[string] += 1 #add a count if match is found 
    
    if string_counter: #if there has been a match, find the keyword with highest count and return it
        return string_counter.most_common(1)[0][0]
    else:
        return None  # Return None if no match is found
    
```

```python
df=pd.read_csv('final_groups.csv')#open csv file of benchmark data 
target_type_list=[]
for aid in df['aids']:
    aid=int(aid)#transform each assay id to an integer 
    t=target_type(str(aid)) #label target type by function 
    target_type_list.append(t) #add to result list 
    
```

```python
# replace some names of target type 
for i in range(len(target_type_list)):
    if target_type_list[i] == "cation":
        target_type_list[i] = "Ion Channel"
    if target_type_list[i]=='protease':
        target_type_list[i] = "Protease"
        
```

```python
num_of_unique_targets=len([*set(target_type_list)]) #find number of target categories
num_of_unique_targets
```

```python
#add results to dataframe and save to csv 
df['Target Type']=target_type_list
df.to_csv('final_groups.csv', index=False)
```
