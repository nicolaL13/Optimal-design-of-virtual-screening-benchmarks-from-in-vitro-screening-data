```python
import pandas as pd
from bs4 import BeautifulSoup
import csv
import requests as req
```

# check each assay to predict if it has a dose response assay 

```python
def is_dose_response(aid):
    """
    function that predicts whether the bioassay has a dose response curve
    
    aid: str PubChem Bioassay unique identifier 
    
    return: Boolean 
    
    """
    string='https://pubchem.ncbi.nlm.nih.gov/rest/pug_view/data/assay/'+aid+'/XML' #add aid to URL 
    # Requesting for the website
    Web = req.get(string,verify=False) 
    # Creating a BeautifulSoup object and specifying the parser
    S = BeautifulSoup(Web.text,features="xml")
    
    results = S.find_all(text=lambda text: text and 'dose-response' in text.lower()) #find all times Dose response was mentioned in xml 
    if len(results)>0: #if found in text, return true 
        return True

    results = S.find_all(text=lambda text: text and 'dose response' in text.lower()) #try different version of dose response to catch other instances 
    if len(results)>0:#if found in text, return true 
        return True
```

```python
#open data into pandas dataframe 
groups_assay=pd.read_csv('final_groups_data.csv')

confirm_assays = groups_assay[groups_assay['type'] == 'confirmatory'] #get list of all confimratory assays 

# Get the values from column 1 for the filtered rows
confirmatory_aid = confirm_assays['AID'].tolist() #get list of all AIDs of confimratory assays 

```

```python
#iterate through each AID and use is_dose_response function. if return True,add aid to list 
dose_response=[]
for aid in confirmatory_aid:
    print(aid)
    if is_dose_response(str(aid)):
        dose_response.append(aid)
        
        
```

```python
def check_dose_response(value):
    """
    function that answers yess or no to whether an assay is a dose response.
    
    value: str 
    return: string 'yes' or 'no' 
    """
    return 'yes' if value in dose_response else 'no'

# Add a new column 'is dose response' to dataframe using check_dose_response function 
groups_assay['is dose response'] = groups_assay['AID'].apply(lambda x: check_dose_response(x))

```

```python
groups_assay.to_csv("dose_response_groups.csv",index=False) #save to csv file to check manually and then add to final groups csv 
```
